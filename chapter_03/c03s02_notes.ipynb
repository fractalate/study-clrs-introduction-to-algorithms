```python
import numpy as np
from scipy.special import gamma

import matplotlib.pyplot as plt
```

```python
# factorial is the gamma function with the argument increased by 1
def factorial(n):
    return gamma(n + 1)
```

## Definitions

* **Polynomially Bounded** - $f(x)$ is polynomial bounded iff $f(x) = O(n^k)$ for some $k > 0$.
* $O(g(n)) = \left\{ f(x) : \text{There exists a constant $c > 0$ and $n_0 > 0$ such that for all $n \ge n_0$ we have $0 \le f(n) \le c g(n)$} \right\}$.

```python
N = 100_000_000_000_000_000
ns = np.arange(1, N + 1, N / 1000)

# the function we're interested in
f1ys = factorial(np.ceil(np.log2(ns)))

# other functions we're comparing against
others = [
    (ns**2.0, 'f(n) = n^2'),
    (ns**3.0, 'f(n) = n^3'),
    (ns**4.0, 'f(n) = n^4'),
]

plt.title('factorial(ceil(lg(n))) vs. Polynomials')
plt.xlabel('n')
plt.ylabel('f(n)')

plt.yscale('log')

plt.plot(ns, f1ys, label='f(n) = factorial(ceil(lg(n)))')
for other, label in others:
    plt.plot(ns, other, label=label)

plt.legend(loc='lower right')

plt.grid(True)
plt.show()
```
